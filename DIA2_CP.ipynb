Elif
Ahora hagamos un programa en el que preguntamis la calificacion del 0 al 100 y queremos saber su equivalente en el sistema de evaluacion estadounidense. Para simplicidad, basemos en este:
A+ = 100
A = 90 - 99
B = 80 - 89
C = 70 - 79
D = 60 - 69
F = < 60

```python
cal = 100
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else: # notemos como al ya no haber condicion, el else cubre todo lo demas
  print("F")
```

```python
# El modulo (%)
# El modulo es un operador matematico que devuelve el residuo de una division
# 10 % 3 da 1, porque 10 dividido enrre 3 cabe 3 y sobra 1
# 7 % 2, porque el 7 no es divisible de forma exacta entre 2
# 2 % 2 ->0
num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El numero {num} es multiplo de 3 o de 5")

# saber si el numero es par o impar usando not
num = 6
if not num % 2 == 0
  print(f"El numero {num} es impar")
```

CICLO "FOR"
los llamados ciclos for, nos ayuda a realizar acciones durantre una cantidad determinada de veces.

```python
# Imprimir los numeros del 0 al 4
for i in range(5): # toma 5 como el numero final pero inicia SIEMPRE en 0
  print(i)
```

```python
for num in range(2,11):
  print(num)
```

```python
# Tmb existe el range(inicio, fin, paso)
for i in range(2,21,2):
  print(i)
```

WHILE

```python
suma = 0
numeros = int(input("Ingresa el numero a suma: "))

while numeros != 0: # != es equivalente a "distinto de ...."
  suma += numeros
  numeros = int(input("Ingresa el numero para sumar o 0 para salir: "))
print(f"Tu suma es: {suma}")
```

```python
# Usar while como for
numero = 0
while numero <= 12:
  print(numero)
  numero += 1
```
