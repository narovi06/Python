Operaciones Matematicas

```python
# 4to Nivel de importancia
# Suma ordinaria
print("Suma ordinaria: ")
print(3 + 4) # 7
variable1,variable2 = 20,40 # 60
print(variable1 + variable2)

# Resta Ordinaria
print("Resta ordinaria: ")
print(2 - 5) # -1
# declaracion de variable
v1 = 10.4
v2 = 15
resultado = v2 - v1
print(resultado) # 4.6
```

```python
# 3er nivel de importancia
# Multiplicacion ordinaria
print("Multiplicacion ordinaria: ")
M1, M2 = 2,3.14
print(M1 * M2) # 6.28

# Division ordinaria
print("--------------------")
print("Division ordinaria: ")
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1 / M2) # 0.4
print(M3 / M4) # 1.
print(M5 / M6) # 1.

# Division de piso
print("--------------------")
print("Division de piso: ")
print(M1 // M2) # 0
print(M3 // M4) # 1
print(M5 // M6) # 0
```

```python
# Potencias
print("Potencias: ")
print(2 ** 3) # 8
P1,P2 = 7,2
print(P1 ** P2) # 49
```

```python
# Area de un circulo
pi = 3.14
r = 4
A = pi * (r ** 2)
print(A)
```
